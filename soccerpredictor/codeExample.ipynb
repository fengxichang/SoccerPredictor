```python
import numpy as np

np.random.seed(42)
n_samples = 1000
target_variable = np.random.choice([1, 5, 9], size=n_samples, p=[0.3, 0.5, 0.2])
```

```python
target_variable
```

```python
from sklearn.utils.class_weight import compute_class_weight
from collections import Counter

cnt_total = {int(k): v for k, v in dict(sorted(Counter(target_variable).items())).items()}
cnt_ratio = {k: v / sum(cnt_total.values()) for k, v in cnt_total.items()}
class_weights = compute_class_weight(class_weight="balanced", classes=np.unique(target_variable), y=target_variable)
class_weights = dict(enumerate(class_weights))

print(f"cnt_total: {cnt_total}")
print(f"cnt_ratio: {cnt_ratio}")
print(f"class_weights: {class_weights}")
```
